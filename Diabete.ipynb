```python

import numpy as np
import pandas as pd
import seaborn as sn

from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import matplotlib as plt
```

```python
data = pd.read_csv(r'Healthcare-Diabetes.csv')
```

```python
data.corr()
```

```python
data.isna().sum()
```

```python
data = data.drop(columns=['Id','BloodPressure','SkinThickness'])
```

```python
data
```

```python
data.corr()
```

```python
X = data.drop(columns='Outcome')
y = data['Outcome']
```

```python
X_train,X_test,y_train,y_text = train_test_split(X,y,train_size=0.8,random_state=43)
```

```python
model = KNeighborsClassifier()
```

```python
kf = KFold(n_splits=20,random_state=43,shuffle=True)

kfscore = cross_val_score(model,X_train,y_train,cv=kf)
```

```python
kfscore
```

```python
#Grid serach

Param_grid = ({'n_neighbors': [1, 3, 5, 7, 9, 11, 15, 25, 50],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'chebyshev', 'minkowski', 'cosine', 'hamming'],
    'p': [1, 2, 3, 4],
    'leaf_size': [10, 20, 30, 40, 50],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'n_jobs': [-1]
    })
```

```python
GridS = RandomizedSearchCV(model,Param_grid,cv=5,scoring='accuracy',n_jobs=-1)
```

```python
GridS
```

```python
GridS.fit(X_train,y_train)
```

```python
GridS.best_params_
```

```python
GridS.best_score_
```

```python
GridS.best_estimator_
```

```python
y_pred = GridS.best_estimator_.predict(X_test)
```

```python
y_pred
```
